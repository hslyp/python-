```
按照示例代码的要求，去尝试补全信贷数据集中的数值型缺失值
```
 1.打开数据（csv文件、excel文件）;
 
 2.查看数据（尺寸信息、查看列名等方法）;
 
 3.查看空值;
 
 4.众数、中位数填补空值;
 
 5.利用循环补全所有列的空值;


````python
import pandas as pd
import numpy as np
data = pd.read_csv("data.csv")

print("数据集行数和列数：",data.shape)
print("数据局列名：",data.columns)
print("数据基本信息：")
print(data.info())
print("数据前五行:",data.head(6))

#查看空值
print("各列空值信息：")
print(data.isnull().sum())
print("缺失值百分比：")
missing_percentage = (data.isnull().sum()/len(data)) * 100
print(missing_percentage)

#众数、中位数填补空值
#利用循环填补空值

for column in data.columns:
    if data[column].dtype == 'object': #对于分类数据（非数值类型），使用众数填补
        mode_value = data[column].mode()[0] #mode()返回该列的众数
        data[column].fillna(mode_value,inplace = True)
        print(f"列 '{column}'使用众数'{mode_value}'填补空值:")
    else: #对于数值型数据，使用中位数填补
        median_value = data[column].median()
        data[column].fillna(median_value,inplace = True)
        print(f"列 '{column}'使用中位数'{median_value}'填补空值")

print("填补空值后的数据：")
print(data.head(6)) 

data.to_csv("data_filled.csv",index = False)
print("填补空值后的数据已保存到data_filled.csv文件中")
````
